```python
# --- Imports ---
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# --- Reproducibility ---
np.random.seed(0)

# --- Load data into a DataFrame ---
iris = load_iris()
feature_names = iris.feature_names

df = pd.DataFrame(iris.data, columns=feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# --- Create a 75/25 train-test split using a boolean mask (as in the tutorial) ---
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75

train = df[df['is_train'] == True].copy()
test  = df[df['is_train'] == False].copy()

print(f"Train size: {len(train)} | Test size: {len(test)}")

# --- Features and labels ---
X_train = train[feature_names]
y_train = train['species']          # strings are fine for scikit-learn classifiers
X_test  = test[feature_names]
y_test  = test['species']

# --- Train Random Forest ---
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)

# --- Evaluate ---
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred, labels=iris.target_names)

print("\nAccuracy:", round(acc, 4))
print("\nConfusion matrix (rows=true, cols=pred):")
print(pd.DataFrame(cm, index=iris.target_names, columns=iris.target_names))

print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# --- Feature importances ---
importances = pd.Series(clf.feature_importances_, index=feature_names).sort_values(ascending=False)
print("\nFeature importances:")
print(importances)

# --- Peek at the data with the new columns ---
display(df.head())
```

```python
# --- Imports ---
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# --- Reproducibility ---
np.random.seed(0)

# --- Load the Iris dataset into a DataFrame ---
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species names as a categorical column
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# --- Create a 75/25 train–test split flag (like in the screenshots) ---
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75

# Split the DataFrame using the boolean mask
train = df[df['is_train'] == True].copy()
test  = df[df['is_train'] == False].copy()

print(f"Number of observations in the training data: {len(train)}")
print(f"Number of observations in the test data: {len(test)}")

# --- Select feature columns (first 4 columns are the numeric features) ---
features = df.columns[:4]
print("\nFeatures:", list(features))

# --- Train a Random Forest on the training set ---
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(train[features], train['species'])

# --- Predict on the test set ---
preds = clf.predict(test[features])

# --- Evaluate ---
acc = accuracy_score(test['species'], preds)
print("\nAccuracy:", round(acc, 4))

cm = confusion_matrix(test['species'], preds, labels=iris.target_names)
cm_df = pd.DataFrame(cm, index=iris.target_names, columns=iris.target_names)
print("\nConfusion matrix (rows=true, cols=pred):")
print(cm_df)

print("\nClassification report:")
print(classification_report(test['species'], preds, target_names=iris.target_names))

# --- Feature importances (most useful features first) ---
importances = (
    pd.Series(clf.feature_importances_, index=features)
      .sort_values(ascending=False)
)
print("\nFeature importances:")
print(importances)

# Quick peek at the top of the prepared DataFrame
print("\nHead of df with species + is_train:")
print(df.head())
```

```python
# --- Pick feature columns (first 4) ---
features = df.columns[:4]

# --- Convert species names to numeric codes (like the tutorial) ---
# y_train are integer codes; uniques holds the species names in the same order
y_train, uniques = pd.factorize(train['species'])

# Make y_test using the SAME ordering as 'uniques'
y_test = pd.Categorical(test['species'], categories=uniques).codes

print("Label mapping:", {name: i for i, name in enumerate(uniques)})

# --- Train Random Forest on numeric targets ---
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(train[features], y_train)

# --- Predict on test set ---
y_pred = clf.predict(test[features])

# --- Evaluate ---
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc = accuracy_score(y_test, y_pred)
print("\nAccuracy:", round(acc, 4))

cm = confusion_matrix(y_test, y_pred, labels=range(len(uniques)))
cm_df = pd.DataFrame(cm, index=uniques, columns=uniques)
print("\nConfusion matrix (rows=true, cols=pred):")
print(cm_df)

print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=list(uniques)))

# --- Optional: feature importances ---
import pandas as pd
importances = (
    pd.Series(clf.feature_importances_, index=features)
      .sort_values(ascending=False)
)
print("\nFeature importances:")
print(importances)
```

```python
 # Step 1: Select features
features = df.columns[:4]

# Step 2: Factorize species into numeric codes for training
y = pd.factorize(train['species'])[0]

# Step 3: Create and train classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[features], y)
```

Predicting on the Test Set, Mapping Predictions back to Species names, and Checking Accuracy

```python
# Predict on the test data
preds = clf.predict(test[features])

# Map numeric predictions back to species names
pred_species = pd.Series(preds).map(lambda x: pd.factorize(train['species'])[1][x])

# Show first 10 predictions alongside actual values
results = pd.DataFrame({
    'Actual': test['species'].values,
    'Predicted': pred_species.values
})
print(results.head(10))

# Evaluate accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Convert actual test species to numeric using the same mapping as training
y_test = pd.Categorical(test['species'],
                        categories=pd.factorize(train['species'])[1]).codes

print("\nAccuracy:", round(accuracy_score(y_test, preds), 4))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, preds))
print("\nClassification Report:")
print(classification_report(y_test, preds, target_names=pd.factorize(train['species'])[1]))
```
